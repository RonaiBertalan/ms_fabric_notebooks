```python
import json

# Check if the Content parameter is a dictionary (as passed by the pipeline) or a raw string.
# In our case, we expect Content to be a dictionary with a key "value" that holds the JSON string.
if isinstance(Content, dict):
    # Ensure the 'value' key exists
    if "value" in Content:
        json_str = Content["value"]
    else:
        raise ValueError("The Content parameter does not have a 'value' key.")
elif isinstance(Content, str):
    json_str = Content
else:
    raise ValueError(f"Unexpected type for Content: {type(Content)}. Expected dict or string.")

# Parse the JSON string into a Python list of dictionaries.
records = json.loads(json_str)

# Create a Spark DataFrame from the list of dictionaries.
df = spark.createDataFrame(records)

# Display the DataFrame.
#display(df)
```

```python
import pandas as pd

# Convert your Spark DataFrame 'df' to a Pandas DataFrame.
pdf = df.toPandas()
```
